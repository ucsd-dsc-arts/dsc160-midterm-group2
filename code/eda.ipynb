# Desciptive Stats

```python
print('{} tracks, {} artists, {} albums, {} genres'.format(
    len(tracks), len(tracks['artist', 'id'].unique()),
    len(tracks['album', 'id'].unique()),
    sum(genres['#tracks'] > 0)))
```

```python
mean_duration = tracks['track', 'duration'].mean()
print('track duration: {:.0f} days total, {:.0f} seconds average'.format(
    sum(tracks['track', 'duration']) / 3600 / 24,
    mean_duration))
```

```python
# Top 20 most popular genres
genres.sort_values(by='#tracks', ascending = False).title.iloc[0:20]
```

# Descriptive Stats for 4 Unique Genres

### Electronic, Rock, Folk, Hip-Hop

```python
tracks['Rock'] = tracks.track.genres_all.apply(lambda x: True if ('12' in x) else False )
tracks['Hiphop'] = tracks.track.genres_all.apply(lambda x: True if ('21' in x) else False )
tracks['Electronic'] = tracks.track.genres_all.apply(lambda x: True if ('15' in x) else False )
tracks['Folk'] = tracks.track.genres_all.apply(lambda x: True if ('17' in x) else False )
```

```python
print('Average Rock Track Duration (in seconds): ', tracks[tracks['Rock'] == True].track.duration.mean().round())
print('Average Hiphop Track Duration (in seconds): ', tracks[tracks['Hiphop'] == True].track.duration.mean().round())
print('Average Electronic Track Duration (in seconds): ', tracks[tracks['Electronic'] == True].track.duration.mean().round())
print('Average Folk Track Duration (in seconds): ', tracks[tracks['Folk'] == True].track.duration.mean().round())
```
