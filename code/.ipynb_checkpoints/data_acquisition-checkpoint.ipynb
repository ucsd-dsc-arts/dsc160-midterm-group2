```python
tuple_index_0 = lambda x: x[0]
```

```python
echo_cols = list(set(list(map(tuple_index_0, echo.columns))))
echo_cols
```

```python
audio_features, metadata, ranks, social_features, temporal_features = pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame(), pd.DataFrame()
```

```python
echo_datasets = [social_features, ranks, temporal_features, metadata, audio_features]
```

```python
for i, col in enumerate(echo_cols):
    print(col)
    echo_datasets[i] = echo[col]
    echo[col].to_csv(os.path.join("..","..", "fma_metadata", "echo_"+str(echo_cols) + ".csv"))
```
